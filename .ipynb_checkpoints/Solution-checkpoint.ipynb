## Import Required Libraries

```python
import random
import torch
import numpy as np
from dqn_agent import Agent
from collections import deque
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm
from unityagents import UnityEnvironment

%matplotlib inline
```

## Load The Environment

```python
env = UnityEnvironment(file_name="/data/Banana_Linux_NoVis/Banana.x86_64")
brain_name = env.brain_names[0]
brain = env.brains[brain_name]
env_info = env.reset(train_mode=False)[brain_name]
state = env_info.vector_observations[0]
action_size = brain.vector_action_space_size
state_size = len(state)
```

## DQN

```python
def dqn(agent, n_episodes=5000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995, train_mode=True,
       print_every=100):
    
    
    scores = []
    scores_window = deque(maxlen=100)
    eps = eps_start
    
    for i_episode in (range(1, n_episodes+1)):
        env_info = env.reset(train_mode=train_mode)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        
        for t in range(max_t):
            action = agent.act(state, eps if train_mode else 0.0)
            env_info = env.step(action)[brain_name]
            next_state = env_info.vector_observations[0]
            reward = env_info.rewards[0]
            done = env_info.local_done[0]
            if train_mode:
                agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            if done:
                break
                
        scores_window.append(score)
        scores.append(score)
        eps = max(eps_end, eps_decay*eps)
        print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)), end="")
        
        if i_episode % print_every == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
        
        if np.mean(scores_window)>=13.0 and train_mode:
            print('\nEnviroment Solved in {} Average Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), 'dqn_checkpoing2.pth')
            break
    return scores
```

## Training

```python
agent = Agent(state_size=state_size, action_size=action_size, seed=0)
```

```python
scores = dqn(agent)
```

```python
score_window = deque(maxlen=100)
mean_scores = []
for s in scores:
    score_window.append(s)
    mean_scores.append(np.mean(score_window))
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(range(len(scores)), scores)
ax.plot(range(len(mean_scores)), mean_scores, c='r', label='Mean score over 100 episodes')
ax.set_xlabel('Episodes')
ax.set_ylabel('Score')
ax.set_title('Score Vs. Episode#')
plt.show()
```
